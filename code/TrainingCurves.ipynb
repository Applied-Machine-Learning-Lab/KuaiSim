```python
%matplotlib inline
from plot_utils import smooth, get_online_training_info, multiplot_multiple_line
import numpy as np

smoothness = 10
param_list = ['0.00001', '0.00003', '0.0001', '0.0003']
param_key = 'LR_'

stats = []
for param in param_list:
    expe = f"ddpg_OneStageHyperPolicy_with_DotScore_actor{param}_critic0.001_niter20000_reg0.00001_ep0.01_noise0.1_bs128_epbs32_step20_seed11"
    log_path = "output/Kuairand_Pure/agents/" + expe  + "/model.report"
    info = get_online_training_info(log_path)

    for k in list(info.keys()):
        v = info[k]
        info[k] = smooth(np.array(v), smoothness)
    print('\t'.join([k for k,v in info.items()]))
    print('\t'.join([str(v[-1]) for k,v in info.items()]))
    stats.append(info)
multiplot_multiple_line([param_key + param for param in param_list], stats, x_name = 'episode')
```
